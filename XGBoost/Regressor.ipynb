# Using regressor model of XGBoost to predict insurance costs

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import root_mean_squared_error,r2_score
from sklearn.preprocessing import OrdinalEncoder
```

```python
data= pd.read_csv('/home/aayam/Documents/MachineLearning/DecisionTrees/insurance.csv')
df= pd.DataFrame(data)

df.head(10)
```

**encoding some categorical features for better predicitions**

```python
encoder= OrdinalEncoder()

df[['sex']]= encoder.fit_transform(df[['sex']])
df[['smoker']]= encoder.fit_transform(df[['smoker']])

df.head(10)
```

**initializing features and targets and predicting**

```python
x= df[['age','sex','bmi','children','smoker']]
y= df['charges']

X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)

model= xgb.XGBRegressor(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    random_state=42
)

cv= cross_val_score(model,X_train,y_train, cv=5)

model.fit(X_train,y_train)

y_pred= model.predict(X_test)

rmse= root_mean_squared_error(y_test,y_pred)
r2= r2_score(y_test,y_pred)

print(f""" rmse: {rmse}\n r2:{r2}\n cv:{cv}\n mean CV:{cv.mean()}\n CV Std: {cv.std()}""")
```
